```python
# info.txt BMI 지수 계산

# bmi = 몸무게 / (키 / 100) ** 2
# bmi >= 25 과체중
# bmi >= 18.5 정상체중
# 저체중


# 출력예시

# 이름: 사나
# 몸무게 : 78
# 키 : 182
# BMI : 12.41225
# 결과 : 저체중

def bmiResult(bmi: float) -> str:
    if bmi >= 25:
        return "과체중"
    elif bmi >= 18.5:
        return "정상체중"
    else:
        return "저체중"

fr = open("info.txt", "r", encoding="utf-8")

data = fr.read()
lines = data.splitlines()

for line in lines:
    line = line.strip()
    if not line:
        continue
    name, whight, height = line.split()
    weight = float(weight)
    height = float(height)

    bmi = weight / (height / 100) ** 2
    result = bmiResult(bmi)

    print(f"이름: {name}")
    print(f"몸무게: {weight:.2f}")
    print(f"키: {height:.2f}")
    print(f"BMI: {bmi:.5f}")
    print(f"결과: {result}")
    print()

fr.close()
```

```python
while True:
    no = int(input("1.암호화 | 2.복호화 | 3.종료 중 원하는 메뉴의 번호 입력 "))

    if no == 1:
        # origin.txt 읽기 => 암호화 후 저장(encryption.txt) + 100
        fr = open("origin.txt","r",encoding="utf-8")
        content = fr.read()
        
        fw = open("encryption.txt","w",encoding="utf-8")
        for c in content:
            fw.write(chr(ord(c) + 100))

        fr.close()
        fw.close()

    elif no == 2:
        # 암호화 파일을 읽은 후 원래 내용 복원시켜서 출력
        fr = open("encryption.txt","r",encoding="utf-8")
        content = fr.read()
        for i in range(len(content)):
            print(chr(ord(content[i]) - 100), end="")

        fr.close()
    else:
        break
```

### json
- loads() : 문자열을 dict 처리
- dumps() : dict 구조를 문자열로
- load() : json => dict
- dump() : dict => json

```python
# json
# {"name": "hog","age":1} ==dict
import json 
data = '{"id":"hong", "language": "python", "edtion":"3.5","author": "Guido van rossum"}'

print(type(data))

#dict => json
# json.loads(str) = > 문자열 json 으로 변환
jsondata = json.loads(data)
print(type(jsondata))

print(jsondata.get('id'))
print(jsondata['id'])

print("dic==> str")
data = {"id":"hong", "language": "python", "edtion":"3.5","author": "Guido van rossum"}
jsondata = json.dumps(data)
print(type(jsondata))

# with open("") as fr:
    
```

```python
# 엑셀 파일
from openpyxl import Workbook ,load_workbook
from random import randint
```

```python
wb = Workbook()

# 첫번째 시트 가져오기
ws = wb.active
# 시트명 변경
ws.title = "test"

# 두번째 시트 생성
ws2 = wb.create_sheet()
ws2.title = "Mysheet"
ws2.sheet_properties.tabColor = "ff66ff"

# 시트 복사
new_ws = wb.copy_worksheet(ws)
new_ws.title = "copy"

wb.save("./data/file1.xlsx")
wb.close()

```
