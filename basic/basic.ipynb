```python
# 사칙연산기 ( 숫자 2, 연산자 => +, -, *, /, //, %)

num1 = int(input('input num1 '))
op = input('input op ')
num2 = int(input('input num2 '))

result = 0
if op == "+":
    result = num1 + num2
elif op == "-":
    result = num1 - num2
elif op == "*":
    result = num1 * num2
elif op == "/":
    result = num1 / num2
elif op == "//":
    result = num1 // num2
elif op == "%":
    result = num1 % num2

print(f"{num1} {op} {num2} = {result}")
```

```python
# 특정 숫자를 입력받은 후 1 ~ 입력받은 숫자까지 합계 구하기

x = int(input("숫자 입력 "))

hap = 0
for i in range(1, x+1):
    hap += i

print(f"1 ~ {x} 까지의 합 {hap}")
```
