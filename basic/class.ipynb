```python
class Car:
    color = ""
    speed = 0

    def upSpeed(self, value):
        self.speed += value

    def downSpeed(self, value):
        self.speed -= value

    def __str__(self):  # == toString()
        return self.color + ":" + str(self.speed)
```

```python
car1 = Car()
car1.color = "Red"
car1.speed = 0

car2 = Car()
car2.color = "Black"
car2.speed = 0

car1.upSpeed(80)
car2.upSpeed(100)

print(car1)
print(car2)
```

```python
# 클래스 메소드
# @classmethod (클래스이름.method()) / @staticmethod (클래스이름.method())
class SelfTest:
    @classmethod
    def func1(cls):  # 클래스 메소드
        print("function1 called")
    
    def func2(self):
        print("function2 called")
```

```python
obj1 = SelfTest()
# obj1.func1()
SelfTest.func1()
obj1.func2()
```

```python
class SelfTest:

    num = 15

    @staticmethod
    def func1(a, b):  # static 메소드 : 변수(클래스변수, self) 사용 안함 / 외부 변수 사용
        return a + b
    
    def func2(self):  # 인스턴스 메소드
        print("function2 called")

    @classmethod
    def func3(cls):  # 클래스 메소드
        print(f"class method {cls.num}")
```

```python
test = SelfTest()
test.func2()

SelfTest.func3()
print(SelfTest.func1(3,5))
```

```python
class Car:

    def __init__(self, color="black", speed=0):
        self.color = color
        self.speed = speed

    def upSpeed(self, value):
        self.speed += value

    def downSpeed(self, value):
        self.speed -= value

    def __str__(self):  # == toString()
        return f"color: {self.color}, speed: {self.speed}"
```

```python
car1 = Car()
print(car1)
car2 = Car(color="white")
print(car2)
car3 = Car(color="black", speed=100)
print(car3)
```

```python
class Parent():
    def __init__(self):
        self.value = "테스트"

    def test(self):
        print("Parent test() 호출")

class Child(Parent):  # 상속
    def __init__(self):
        Parent.__init__(self)  # == super() 부모의 인스턴스 변수 호출여부
        print("Child 클래스의 __init__호출")
```

```python
parent = Parent()
parent.test()

child = Child()
child.test()
print(child.value)  # 부모의 변수 호출
```

```python
class AudioVisual:
    def __init__(self, power, volume):
        self.power = power
        self.volume = volume

    def switch(self, onOff):
        self.power = onOff
    
    def setVolume(self, volume):
        self.volume = volume


class Audio(AudioVisual):
    def __init__(self, power, volume):
        super().__init__(power, volume)

    def tune(self):
        str = "La la la...." if self.power else "turn it on"
        print(str)


class Tv(AudioVisual):
    def __init__(self, power, volume, size):
        super().__init__(power, volume)
        self.size = size
    
    def watch(self):
        str = "have fun!!" if self.power else "switch on"
        print(str)
```

```python
obj1 = Tv(False, 12, 40)
obj1.switch(True)
obj1.watch()

obj2 = Audio(False, 15)
obj2.setVolume(6)
obj2.tune()
```

```python
# 오버라이딩

class Parent():
    def __init__(self):
        self.value = "테스트"

    def func1(self):
        print("Parent test() 호출")

    def func2(self):
        print("Parent func2() 호출")


class Child(Parent):  # 상속
    def __init__(self):
        Parent.__init__(self)  # == super() 부모의 인스턴스 변수 호출여부
        print("Child 클래스의 __init__호출")

    def func1(self):
        print("Parent func1() 재정의")

    def func3(self):
        print("Child func3() 호출")
```

```python
parent, child = Parent(), Child()
parent.func1()
parent.func2()

child.func1()
child.func2()
child.func3()
```

```python
# private 키워드는 없음
# __변수 : private 으로 취급됨

import math
class Circle():
    def __init__(self, radius):
        self.__radius = radius

    def getcirc(self):
        return 2 * math.pi * self.__radius
    
    def getarea(self):
        return math.pi * (self.__radius ** 2)
    
    def getradius(self):
        return self.__radius
    
    def setradius(self, radius):
        self.__radius = radius
```

```python
circle = Circle(10)
print(f"둘레 {circle.getcirc()}")
print(f"면적 {circle.getarea()}")

# print(circle.__radius) # AttributeError: 'Circle' object has no attribute '__radius'
```
